```python
#7 Replace missing values or outliers with any mean or interpolated values.


TDC = int(df["total day calls"].mean())
print(TDC)
df["total day calls"].fillna(TDC,inplace=True)

TEC=int(df["total eve calls"].mean())
print(TEC)
df["total eve calls"].fillna(TEC,inplace=True)

TNC=int(df["total night calls"].mean())
print(TNC)
df["total night calls"].fillna(TNC,inplace=True)

TIC=int(df["total intl calls"].mean())
print(TIC)
df["total intl calls"].fillna(TIC,inplace=True)

print("Replaced missing values by Adding Mean Value")
df.isnull().sum()
```

```python
#12 Display a bar chart of number of churns happened with total eve calls > 100

import matplotlib.pyplot as plt
totalcalls=pd.DataFrame(df,columns=["total eve calls","churn"])
ds = totalcalls.loc[totalcalls["total eve calls"]>100]
print(ds)

plt.bar(x=ds["churn"],height=ds["total eve calls"])
plt.title("Churns VS Total eve calls")
plt.ylabel("TOTAL EVE CALLS")
plt.xlabel("CHURNS")
plt.show()
```
